# Seaborn!
Seaborn is an advance "Data-Visualization" Library, that uses "Matplotlib" to plot graph in the background! Its graphs are pritter and more advance then "Matplotlib"(overall)!

##### Modules:
To use "seaborn" we have to import "matplotlib as well"! Since "seaborn" requires matplotlib to be run in background!

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2,500)
y = np.cos(x/3*4)
sns.set()   # This is the man line of code which indicates the computer to use seaborn instead of "matplotlib"
plt.plot(x,y)
plt.show()
```

## Conclusion!
"seaborn" includes all the "matplotlib" functions and graph! Make the whole code for lot just like we have done in "matplotlib" and just mention "sns.set()" in the code to get the final result is "seaborn"!
